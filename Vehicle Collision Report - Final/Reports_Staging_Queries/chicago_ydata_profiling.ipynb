```python
# Import the ProfileReport class from the ydata_profiling package
from ydata_profiling import ProfileReport

# Import the pandas library under the alias pd
import pandas as pd
```

```python
# Read the TSV file into a pandas DataFrame named df_chicago
# The file is tab-separated (sep='\t')
# Set low_memory=False to ensure that the entire file is read into memory at once (may be necessary for large files)
df_chicago = pd.read_csv("Chicago_Traffic_Crashes_-_Crashes_20240326.tsv", sep='\t', low_memory=False)

# Display the first few rows of the DataFrame to inspect the data
df_chicago.head()
```

```python
# Generate a data profiling report for the DataFrame df_chicago
# The title parameter sets the title of the report to "Chicago Traffic Crashes"
profile_chicago = ProfileReport(df_chicago, title = "Chicago Traffic Crashes")
```

```python
# Save the data profiling report to an HTML file named "Chicago_Traffic_Crashes.html"
profile_chicago.to_file("Chicago_Traffic_Crashes.html")
```
